```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./administration_data.csv')
```

```python
df
```

```python
df_m = df[df.gender == 'Male'].groupby('age').sum()
df_fm = df[df.gender == 'Female'].groupby('age').sum()
```

```python
plt.figure(figsize = (16,9))

plt.title('График зависимости возраста и количества воспринимаемых и порождаемых слов', fontsize=20)

plt.plot(df_m.index, df_m.comprehension/len(df_m), '--r', label='Восприятие, мальчики')
plt.plot(df_m.index, df_m.production/len(df_m), ':b', label='Порождение, мальчики')
plt.plot(df_fm.index, df_fm.comprehension/len(df_fm), '--y', label='Восприятие, девочки')
plt.plot(df_fm.index, df_fm.production/len(df_fm), ':g', label='Порождение, девочки')

plt.axvline(x=16, color='black', alpha=0.5)
plt.axvline(x=24, color='black', alpha=0.5)
plt.axvline(x=32, color='black', alpha=0.5)

plt.gca().yaxis.grid(alpha=0.3)

plt.xlabel('Возраст ребёнка', fontsize=17)
plt.ylabel('Количество слов', fontsize=17)

plt.legend()
plt.show()
```

#### Слияние восприятия и порождения слов наступает в возрасте 19 месяцев
